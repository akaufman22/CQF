```python
import numpy as np
import matplotlib.pyplot as plt
from binomial import binomial_option
```

```python
S = 100
K = 100
r = 0.05
T = 1
N = 4
```

```python
vols = np.linspace(0, 1.2, 121)
values = np.array([binomial_option(S,K,r,v,T,N,2)[0,0] for v in vols])
```

```python
plt.plot(vols, values)
```

```python
st = range (1,200)
values = np.array([binomial_option(S,K,r,0.2,T,s,2)[0,0] for s in st])
```

```python
plt.plot(st, values)
```
